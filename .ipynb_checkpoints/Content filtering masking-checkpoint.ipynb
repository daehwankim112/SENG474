```python
import pandas as pd
import numpy as np
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
```

```python
# 1. Load the similar playlist id
all_df_pids = pd.read_csv("all_similar_pids.csv")
df_pid = pd.read_csv("similar_pids.csv")
df_nf = pd.read_csv("new_features.csv")
pid_list = df_pid.iloc[:, 0]
#pid_list = all_df_pids["pid"]
print(pid_list)
print(all_df_pids)
print(all_df_pids.iloc[0, :])
```

```python
all_songs_numbers_only = df_nf.drop(columns=['pid', 'track_uri', 'track_name', 'artist_name', 'id', 'name', 'album', 'album_id', 'artists', 'artist_ids', 'track_number', 'disc_number', 'explicit', 'duration_ms', 'time_signature', 'year', 'release_date'], errors='ignore').drop_duplicates().reset_index(drop=True)
print(all_songs_numbers_only.shape)
```

```python
filtered_songs = df_nf[df_nf["pid"].isin(pid_list)].drop(columns=['pid'], errors='ignore').drop_duplicates().reset_index(drop=True)
# print(filtered_songs)
print(filtered_songs.shape)
```

```python
all_pid_filtered_songs = []
pid_count = 0
total_pid_count = all_df_pids.shape[0]

#for pid_from_all_list in all_df_pids["similar_pids"]:
for i in range(total_pid_count):
    
    pid_count += 1
    if divmod(pid_count, 1000)[1] == 0:
        print("{:.2f}".format(pid_count / total_pid_count * 100), "% done")

    # print("pid_from_all_list", pid_from_all_list)
    
    # ✅ Wrap `pid_from_all_list` in a list
    all_filtered_songs = df_nf[df_nf["pid"].isin([all_df_pids["similar_pids"].iloc[i]])].drop(
        columns=['pid'], errors='ignore'
    ).drop_duplicates().reset_index(drop=True)

    all_pid_filtered_songs.append((all_df_pids["pid"].iloc[i], filtered_songs))
```

```python
combined_df = pd.concat(
    [df.assign(pid=pid) for pid, df in all_pid_filtered_songs], ignore_index=True
)
combined_df.to_csv("all_pid_filtered_songs.csv", index=False)

print("✅ Exported all_pid_filtered_songs to all_pid_filtered_songs.csv")
```

```python
filtered_songs_numbers_only = filtered_songs.drop(columns=['track_id', 'track_uri', 'track_name', 'artist_name', 'id', 'name', 'album', 'album_id', 'artists', 'artist_ids', 'track_number', 'disc_number', 'explicit', 'duration_ms', 'time_signature', 'year', 'release_date'], errors='ignore')
# print(filtered_songs_numbers_only)
#filtered_songs_numbers_only_dict = {filtered_songs.loc[i, 'track_id']: filtered_songs_numbers_only[i] for i in range(len(filtered_songs))}
filtered_songs_numbers_only_dict = {
    filtered_songs.loc[i, 'track_id']: filtered_songs_numbers_only.iloc[i].to_dict()
    for i in range(len(filtered_songs))
}
# print(filtered_songs_numbers_only_dict)


#filtered_songs_numbers_only_dict = {
#    row['track_id']: row.drop(labels=['track_id']).to_dict()
#    for _, row in filtered_songs.iterrows()
#}
```

```python
all_filtered_songs_numbers_only = all_filtered_songs.drop(columns=['track_id', 'track_uri', 'track_name', 'artist_name', 'id', 'name', 'album', 'album_id', 'artists', 'artist_ids', 'track_number', 'disc_number', 'explicit', 'duration_ms', 'time_signature', 'year', 'release_date'], errors='ignore')

all_filtered_songs_numbers_only_dict = {
    row['track_id']: row.drop(labels=['track_id']).to_dict()
    for _, row in all_filtered_songs.iterrows()
}
```

```python
# 2. Train K Nearest Neighbors model
nbrs = NearestNeighbors(n_neighbors=10, algorithm='ball_tree')
all_track_ids_list = list(filtered_songs_numbers_only_dict.keys())
all_vectors = np.array([list(v.values()) for v in filtered_songs_numbers_only_dict.values()], dtype=float)
# nbrs.fit(all_vectors, all_track_ids_list)
nbrs.fit(all_vectors)
```

```python
# Dictionary to store trained KNN models for each playlist
playlist_knn_models = {}

# Iterate over all filtered playlist songs
for pid, df in all_pid_filtered_songs:
    if df.empty:
        print(f"Skipping empty playlist {pid}")
        continue  # Skip empty playlists

    # Extract track features (numeric values only)
    feature_df = df.drop(columns=['track_id', 'track_uri', 'track_name', 'artist_name', 'id', 'name', 
                                  'album', 'album_id', 'artists', 'artist_ids', 'track_number', 
                                  'disc_number', 'explicit', 'duration_ms', 'time_signature', 
                                  'year', 'release_date'], errors='ignore')
    
    # Convert to NumPy array
    all_vectors = np.array(feature_df, dtype=float)

    # Initialize and train the model
    knn = NearestNeighbors(n_neighbors=10, algorithm='ball_tree')
    knn.fit(all_vectors)
    
    # Store the trained model
    playlist_knn_models[pid] = knn
```

```python
def recommend_nearest(target_track_id, pid_in, top_n=10):
    if target_track_id not in all_songs_numbers_only['track_id'].str.strip().values:
        print(f"Track {target_track_id} not found.")
        return []
    
    if pid_in not in playlist_knn_models:
        print(f"Warning: No KNN model found for pid {pid_in}")
        return []
    
    target_vector = all_songs_numbers_only[all_songs_numbers_only['track_id'] == target_track_id].iloc[0, 1:].values.reshape(1, -1)
    # target_vector = np.array(all_songs_numbers_only['track_id'==target_track_id]).reshape(1, -1)
    distances, indices = playlist_knn_models[pid_in].kneighbors(target_vector)
    recommendations = [
        (all_track_ids_list[idx], distances[0][j]) 
        for j, idx in enumerate(indices[0]) 
        if distances[0][j] > 0  # Only include songs with a distance greater than 0
    ]
    return recommendations[:top_n]
```

```python
pid_count = 0

for pid in all_df_pids["pid"]:
    pid_count += 1
    if divmod(pid_count, 400)[1] == 0:
        print("{:.2f}".format(pid_count / total_pid_count * 100), "% done")
    
    if pid not in playlist_knn_models:
        print(f"❌ Skipping playlist {pid}, not found in playlist_knn_models")
        continue  # Skip this playlist

    recommendation_songs_rankings = [
        recommend_nearest(track_id, pid, 1)[0]  # Extract first element
        for track_id in df_nf[df_nf["pid"] == 0]["track_id"][:-1]  # masking
    ]
```

```python
# Create an empty list to store recommended songs
all_recommended_songs = []
playlists_with_one_song_pid =[]
masked_answers = []

# To display progress
len_of_pids = len(all_df_pids["pid"].unique())
pid_counter = 0

# Iterate over all unique pid values
for pid in all_df_pids["pid"].unique():
    # Get track_ids for the current pid, excluding the last one
    track_ids = df_nf[df_nf["pid"] == pid]["track_id"][:-1]
    
    
    # Generate the recommendation rankings for the current pid
    recommendation_songs_rankings = [
        recommend_nearest(track_id, pid, 1)[0]  # Extract the first element (recommended song)
        for track_id in track_ids
    ]
    
    # Only proceed if we have valid recommendations
    if recommendation_songs_rankings:
        # Sort by distance and get the closest recommended song
        recommended_song = sorted(recommendation_songs_rankings, key=lambda x: x[1])[0][0]
        # Append the recommended song for this pid to the list
        all_recommended_songs.append((pid, recommended_song))

        masked_answers.append([df_nf[df_nf["pid"] == pid]["track_id"].iloc[-1]])
    else:
        # print(f"No recommendations found for pid {pid}")
        playlists_with_one_song_pid.append(pid)

    # report progress of this algorithm
    pid_counter += 1
    if divmod(pid_counter, 800)[1] == 0:
        print("{:.2f}".format(pid_counter / len_of_pids * 100), "% done")

# Print all recommended songs (one per pid)
#for pid, recommended_song in all_recommended_songs:
#    print(f"Recommended song for pid {pid}: {recommended_song}")

print("Number of playlist with one songs:", len(playlists_with_one_song_pid))
print("Number of recommended songs:", len(all_recommended_songs))
```

```python
# Export all_recommended_songs to CSV
recommended_songs_df = pd.DataFrame(all_recommended_songs, columns=["pid", "recommended_song"])
recommended_songs_df.to_csv("all_recommended_songs.csv", index=False)

# Export masked_answers to CSV
masked_answers_df = pd.DataFrame(masked_answers, columns=["masked_answer"])
masked_answers_df.to_csv("all_masked_answers.csv", index=False)
```
